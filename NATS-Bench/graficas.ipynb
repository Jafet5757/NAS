```python
import matplotlib.pyplot as plt
import numpy as np

# definimos los vectores a graficar
time_ciafr10 = [44.20, 45.10, 193.00]
time_ciafr100 = [37.76, 36.91, 200.98]

pearson_cifar10 = [0.82, 0.68, 0.57]
pearson_cifar100 = [0.79, 0.53, 0.49]

spearman_cifar10 = [0.61, 0.54, 0.58]
spearman_cifar100 = [0.63, 0.45, 0.44]
```

```python
# hacemos un grafico de barras para los tiempos (superpuestas en cifar10 y cifar100)
fig, ax = plt.subplots()
bar_width = 0.35
index = np.arange(3)
opacity = 0.8

rects1 = plt.bar(index, time_ciafr10, bar_width, alpha=opacity, color='b', label='CIFAR-10')
rects2 = plt.bar(index + bar_width, time_ciafr100, bar_width, alpha=opacity, color='orange', label='CIFAR-100')

plt.xlabel('Método')
plt.ylabel('Tiempo (s)')
plt.title('Tiempo de ejecución de los métodos')
plt.xticks(index + bar_width, ('Propuesta final', 'Synflow', 'Similitud gradiente'))
plt.legend()
```

```python
# Graficamos los coeficientes de Pearson
plt.plot(pearson_cifar10, 'o-', label='CIFAR-10', color='b')
plt.plot(pearson_cifar100, 'o-', label='CIFAR-100', color='orange')
plt.xlabel('Método')
plt.ylabel('Coeficiente de Pearson')
plt.title('Coeficiente de Pearson de los métodos')
plt.xticks(index, ('Propuesta final', 'Synflow', 'Similitud gradiente'))
plt.legend()
```

```python
# Graficamos los coeficientes de Spearman
plt.plot(spearman_cifar10, 'o-', label='CIFAR-10', color='b')
plt.plot(spearman_cifar100, 'o-', label='CIFAR-100', color='orange')
plt.xlabel('Método')
plt.ylabel('Coeficiente de Spearman')
plt.title('Coeficiente de Spearman de los métodos')
plt.xticks(index, ('Propuesta final', 'Synflow', 'Similitud gradiente'))
plt.legend()
```
